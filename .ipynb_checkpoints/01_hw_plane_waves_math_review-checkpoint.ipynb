**<span style="color:#ffaaff;">Approximations Review</span>**

In this course, you will need to remember and use some basic approximations.\
These approximations all come from taking the [*Taylor Expansion*](https://en.wikipedia.org/wiki/Taylor_series) of a function $f(x)$ about some point $x = a$:
$$f(x)\Bigr|_{x \rightarrow a} \approx f(a) + f'(a) (x-a) + \dfrac{1}{2!} f''(a) (x-a)^2 = \displaystyle \sum_{n=0}^\infty \dfrac{f^{(n)}(a)}{n!}(x - a)^n
$$

## **Question 1: Binomial Approximation**

</div>

**The binomial approximation to first order in $x$ is as follows:**
\begin{align}
(1 + x)^{n} \approx 1 + n x.
\end{align}

<div style="background-color: #98C1D9; padding: 3px; border-radius: 5px;">

### **Question 1A Solution** (Already filled out)

</div>

**Derive the binomial approximation using the Taylor Expansion to first order about $x = 0$**

---
  
Let $$f(x) = (1 + x)^n,$$ 
then at $x = 0$, $$f(0) = 1.$$
Then the first derivative $f'(x)$ is $$f'(x) = n (1 + x)^{n-1}$$
and the derivative evaluated at $x = 0$ is $$f'(0) = n$$

The Taylor Expansion to first order then becomes 
\begin{align}
f(x)\Bigr|_{x \rightarrow 0} &\approx f(0) + f'(0) x\\~\\
&= \boxed{1 + n x}
\end{align}

<div style="background-color: #98C1D9; padding: 3px; border-radius: 5px;">

### **Question 1B Solution**

</div>

**Find the second and third order terms of the binomial approximation**  

---
  
Taking the second and third derivatives, and evaluating at 0 yields
\begin{align}
f''(x) &= n (n-1) (1 + x)^{n-2} \\
f'''(x) &= n (n-1) (n-2) (1 + x)^{n-3} \\~\\
f''(0) &= n (n-1)\\
f''(0) &= n (n-1) (n-2)
\end{align}

The second order expansion is
\begin{align}
f(x)\Bigr|_{x \rightarrow 0} &\approx f(0) + f'(0) x + \dfrac{1}{2!} f''(0) x^2\\~\\
&= \boxed{1 + n x + \dfrac{1}{2} n (n-1) x^2}
\end{align}

The third order expansion is
\begin{align}
f(x)\Bigr|_{x \rightarrow 0} &\approx f(0) + f'(0) x + \dfrac{1}{2!} f''(0) x^2 + \dfrac{1}{3!} f'''(0) x^3\\~\\
&= \boxed{
1 + n x + \dfrac{1}{2} n (n-1) x^2 + \dfrac{1}{6} n (n-1) (n-2) x^3
}
\end{align}


<div style="background-color: #98C1D9; padding: 3px; border-radius: 5px;">

### **Question 1C Solution**

</div>

* **Plot the binomial function on $x \in [-1, 1]$ for $n = \dfrac{1}{2}$.**
* **Compare to plots of the first, second, and third order binomial approximation.**
* **At what $x > 0$ does each approximation fail, becoming greater than 5\% error?**
---

```python
def binom(xx:float, nn:float):
    """Binomial function (1 + xx)^nn
    
    Inputs:
    -------
    xx: float or array of floats
        binomial variable
    nn: float
        binomial exponent

    Output:
    -------
    binom: float or array of floats
        binomial expansion
    """
    return (1 + xx)**nn
```

```python
# Parameter definitions.  Protip: never make single-letter variable names
nn = 0.5
xx = np.linspace(-1, 2, 100)

taylor0 = 1
taylor1 = taylor0 + nn * xx
taylor2 = taylor1 + 0.5 * nn * (nn - 1) * xx**2
taylor3 = taylor2 + (1/6) * nn * (nn - 1) * (nn - 2) * xx**3
```

```python
# At which x does the error become greater than 10%?
# First, we divide the approximation by the real function,
# Second, we subtract 1 from that ratio
# Third, we take the absolute value of the subtraction
# Fourth, we look for the first location where the final result is greater than 0.1
# Fifth, we find where x > 0
# Sixth, we take the intersection of the indices found
# Seventh, we find the first index where the error is large for plotting
error = 0.05
model = binom(xx, nn)

xx_errors = np.array([])
for taylor in [taylor1, taylor2, taylor3]:
    abs_errors = np.abs(taylor/model - 1) # final result
    indices_error = np.argwhere(abs_errors > error)
    indices_x = np.argwhere(xx > 0)
    
    indices_final = np.intersect1d(indices_error, indices_x)
    index = indices_final[0]
    
    xx_errors = np.append(xx_errors, xx[index])
print(xx_errors)
```

```python
# my linux distro really doesn't like the way tex is used here so i removed it
fig, s1 = plt.subplots(1)

s1.plot(xx, binom(xx, nn), label="Binomial Function")
s1.plot(xx, taylor1, ls="--", label="Taylor 1")
s1.plot(xx, taylor2, ls="--", label="Taylor 2")
s1.plot(xx, taylor3, ls="--", label="Taylor 3")

s1.axvline(x=0, label="x = 0")

for ii, xx_error in enumerate(xx_errors):
    s1.axvline(x=xx_error, color=f"C{ii+1}", ls=":", label=f"Taylor {ii+1} Error")

s1.set_title(f"Binomial Approximations about x = 0 for n = {nn}")
s1.set_xlabel("x")
s1.set_ylabel("f(x)")

s1.legend()
s1.grid()
plt.show()
```

**The approximations fail at the following:**  
1. First Order: $\boxed{0.879}$
2. Second Order: $\boxed{1.303}$
3. Third Order: $\boxed{1.424}$

### **Question 2A: Sine**

</div>

**Repeat the Taylor Expansion approximations for sine  to second order about $x = 0$.**  
**Make the plots, but you don't need to calculate the 5\% error point.**  

---

Recall the Taylor expansion formula:

$$
f(x)\Bigr|_{x \rightarrow 0} &\approx f(0) + f'(0) x + \dfrac{1}{2!} f''(0) x^2 + \dfrac{1}{3!} f'''(0) x^3 +\cdots
$$
  
For $f(x) = \sin(x)$, compute the derivatives:

$$
\begin{aligned}
f(x) &= \sin(x) &\implies f(0) &= \sin(0) = 0\\
f'(x) &= \cos(x) &\implies f'(0) &= \cos(0) = 1\\
f''(x) &= -\sin(x) &\implies f''(0) &= -\sin(0) = 0\\
f'''(x) &= -\cos(x) &\implies f'''(0) &= -\cos(0) = -1\\
\end{aligned}
$$

Now plug into the Taylor formula:

$$
\begin{aligned}
\text{First order:} \quad & \sin(x) \approx f(0) + f'(0)x = 0 + 1 \cdot x = \boxed{x} \\[1mm]
\text{Second order:} \quad & \sin(x) \approx x + \frac{f''(0)}{2!}x^2 = x + 0 = \boxed{x} \\[1mm]
\text{Third order:} \quad & \sin(x) \approx x + \frac{f''(0)}{2!}x^2 + \frac{f'''(0)}{3!}x^3 = x + 0 - \frac{1}{6}x^3 = \boxed{x - \frac{x^3}{6}}
\end{aligned}
$$


```python
def sine(xx):
    """Sine function
    
    Inputs:
    -------
    xx : float or array of floats

    Output:
    -------
    sin(xx)
    """
    return np.sin(xx)

# domain
xx = np.linspace(-2*np.pi, 2*np.pi, 400)

#expansions
taylor1 = xx
taylor2 = taylor1 #same thing
taylor3 = taylor2 - (1/6) * xx**3

# plot
fig, s1 = plt.subplots(1)

s1.plot(xx, sine(xx), label="sin(x)")
s1.plot(xx, taylor1, ls="--", label="Taylor 1")
s1.plot(xx, taylor2, ls="--", label="Taylor 2")
s1.plot(xx, taylor3, ls="--", label="Taylor 3")

# expanding about 0
s1.axvline(x=0, label="x = 0")
s1.set_xlim(-4, 4)
s1.set_ylim(-2, 2)

s1.set_title("Sine Approximations about x = 0")
s1.set_xlabel("x")
s1.set_ylabel("f(x)")

s1.legend()
s1.grid()
plt.show()
```

### **Question 2B: Cosine**

</div>

**Repeat the Taylor Expansion approximations for cosine to second order about $x = 0$.**  
**Make the plots, but you don't need to calculate the 5\% error point.**  

---

For $f(x) = \cos(x)$, compute the derivatives:

$$
\begin{aligned}
f(x) &= \cos(x) &\implies f(0) &= \cos(0) = 1\\
f'(x) &= -\sin(x) &\implies f'(0) &= -\sin(0) = 0\\
f''(x) &= -\cos(x) &\implies f''(0) &= -\cos(0) = -1\\
f'''(x) &= \sin(x) &\implies f'''(0) &= \sin(0) = 0\\
\end{aligned}
$$

Now plug into the Taylor formula:

$$
\begin{aligned}
\text{First order:} \quad & \cos(x) \approx f(0) + f'(0)x = 1 + 0 = \boxed{1} \\[1mm]
\text{Second order:} \quad & \cos(x) \approx 1 + \frac{f''(0)}{2!}x^2 = 1 - \frac{1}{2}x^2 = \boxed{1 - \frac{x^2}{2}} \\[1mm]
\text{Third order:} \quad & \cos(x) \approx 1 - \frac{x^2}{2} + \frac{f'''(0)}{3!}x^3 = 1 - \frac{x^2}{2} + 0 = \boxed{1 - \frac{x^2}{2}}
\end{aligned}
$$

```python
def cosine(xx):
    """Sine function
    
    Inputs:
    -------
    xx : float or array of floats

    Output:
    -------
    sin(xx)
    """
    return np.cos(xx)

# domain
xx = np.linspace(-2*np.pi, 2*np.pi, 400)

#expansions
taylor1 = xx
taylor2 = 1 - 0.5 * xx**2
taylor3 = taylor2 #same thing

# plot
fig, s1 = plt.subplots(1)

s1.plot(xx, cosine(xx), label="cos(x)")
s1.plot(xx, taylor1, ls="--", label="Taylor 1")
s1.plot(xx, taylor2, ls="--", label="Taylor 2")
s1.plot(xx, taylor3, ls="--", label="Taylor 3")

# expanding about 0
s1.axvline(x=0, label="x = 0")
s1.set_xlim(-4, 4)
s1.set_ylim(-2, 2)

s1.set_title("Cosine Approximations about x = 0")
s1.set_xlabel("x")
s1.set_ylabel("f(x)")

s1.legend()
s1.grid()
plt.show()
```

### **Question 3B:**

</div>

**Calculate the magnitude $r(\phi)$ and argument $\theta(\phi)$ for each $z_i$.**  

---  

1. $z_1(\phi) = 2 + e^{i\phi}$:

\begin{align}
\Re(z_1) &= 2 + \cos\phi, \quad \Im(z_1) = \sin\phi \\
\\
r_1(\phi) &= |z_1(\phi)| = \sqrt{ (\Re(z_1))^2 + (\Im(z_1))^2 } \\
          &= \sqrt{ (2 + \cos\phi)^2 + (\sin\phi)^2 } \\
          &= \sqrt{ 4 + 4\cos\phi + \cos^2\phi + \sin^2\phi } \\
          &= \sqrt{ 4 + 4\cos\phi + 1 } \quad (\cos^2\phi + \sin^2\phi = 1) \\
          &= \sqrt{ 5 + 4\cos\phi } \\
          &= \boxed{\sqrt{5 + 4\cos\phi}} \\
\\
\theta_1(\phi) &= \arg(z_1(\phi)) = \arctan\left( \frac{\Im(z_1)}{\Re(z_1)} \right) \\
                &= \arctan\left( \frac{\sin\phi}{2 + \cos\phi} \right) \\
                &= \boxed{\arctan\left( \frac{\sin\phi}{2 + \cos\phi} \right)}
\end{align}

---

2. $z_2(\phi) = \frac{3}{2 - e^{i\phi}}$:

\begin{align}
2 - e^{i\phi} &= 2 - (\cos\phi + i \sin\phi) = (2 - \cos\phi) - i \sin\phi \\
\\
z_2(\phi) &= \frac{3}{(2 - \cos\phi) - i \sin\phi} 
          = \frac{3((2 - \cos\phi) + i \sin\phi)}{(2 - \cos\phi)^2 + (\sin\phi)^2} \\
\\
\Re(z_2) &= \frac{3 (2 - \cos\phi)}{(2 - \cos\phi)^2 + (\sin\phi)^2}, \quad
\Im(z_2) = \frac{3 (\sin\phi)}{(2 - \cos\phi)^2 + (\sin\phi)^2} \\
\\
r_2(\phi) &= \sqrt{ (\Re(z_2))^2 + (\Im(z_2))^2 } \\
          &= \sqrt{ \frac{9 (2 - \cos\phi)^2}{((2 - \cos\phi)^2 + (\sin\phi)^2)^2} + \frac{9 (\sin\phi)^2}{((2 - \cos\phi)^2 + (\sin\phi)^2)^2} } \\
          &= \frac{\sqrt{9 ((2 - \cos\phi)^2 + (\sin\phi)^2)}}{(2 - \cos\phi)^2 + (\sin\phi)^2} \\
          &= \frac{3}{\sqrt{(2 - \cos\phi)^2 + (\sin\phi)^2}} \\
          &= \frac{3}{\sqrt{4 - 4\cos\phi + \cos^2\phi + \sin^2\phi}} \\
          &= \frac{3}{\sqrt{5 - 4\cos\phi}} \\
          &= \boxed{\frac{3}{\sqrt{5 - 4\cos\phi}}} \\
\\
\theta_2(\phi) &= \arg(z_2(\phi)) = \arctan\left( \frac{\Im(z_2)}{\Re(z_2)} \right) \\
                &= \arctan\left( \frac{\sin\phi}{2 - \cos\phi} \right) \\
                &= - \arctan\left( \frac{\sin\phi}{2 - \cos\phi} \right) \\
                &= \boxed{- \arctan\left( \frac{\sin\phi}{2 - \cos\phi} \right)}
\end{align}

---

3. $z_3(\phi) = e^{(\sigma + i\omega)\phi}, \ \sigma = -0.5, \ \omega = 1$:

\begin{align}
z_3(\phi) &= e^{\sigma \phi} e^{i \omega \phi} = e^{\sigma \phi} (\cos(\omega \phi) + i \sin(\omega \phi)) \\
\Re(z_3) &= e^{\sigma \phi} \cos(\omega \phi) = e^{-0.5 \phi} \cos\phi, \quad
\Im(z_3) = e^{\sigma \phi} \sin(\omega \phi) = e^{-0.5 \phi} \sin\phi \\
\\
r_3(\phi) &= |z_3(\phi)| = \sqrt{ (\Re(z_3))^2 + (\Im(z_3))^2 } \\
          &= \sqrt{ e^{-1\phi} (\cos^2\phi + \sin^2\phi) } \\
          &= e^{-0.5 \phi} \\
          &= \boxed{e^{-0.5 \phi}} \\
\\
\theta_3(\phi) &= \arg(z_3(\phi)) = \arctan\left( \frac{\Im(z_3)}{\Re(z_3)} \right) \\
                &= \arctan\left( \frac{e^{-0.5 \phi} \sin\phi}{e^{-0.5 \phi} \cos\phi} \right) \\
                &= \arctan(\tan\phi) = \phi \\
                &= \boxed{\phi}
\end{align}

### **Question 3C:**

**Calculate the velocity of the phasors with respect to $\phi$, and draw them for each $z_i$ evaluated at $\phi = \left\{0, \dfrac{\pi}{2}, \pi, \dfrac{3\pi}{2} \right\}$**  

---
  
1. $z_1(\phi) = 2 + e^{i\phi}$

\begin{align}
v_1(\phi) &= \frac{dz_1}{d\phi} = \frac{d}{d\phi} \left(2 + e^{i\phi}\right) = i e^{i\phi} \\
\\

|v_1| &= \sqrt{ (-\sin\phi)^2 + (\cos\phi)^2 } = \sqrt{\sin^2\phi + \cos^2\phi} = \boxed{1} \\
\arg(v_1) &= \arctan\left( \frac{\Im(v_1)}{\Re(v_1)} \right) = \arctan\left( \frac{\cos\phi}{-\sin\phi} \right)
\end{align}

At the specified points:

\begin{align*}
\phi = 0: &\quad v_1 = i = (0,1) \\
\phi = \frac{\pi}{2}: &\quad v_1 = -1 = (-1,0) \\
\phi = \pi: &\quad v_1 = -i = (0,-1) \\
\phi = \frac{3\pi}{2}: &\quad v_1 = 1 = (1,0)
\end{align*}

```python
plt.close('all')
# specified phi values
phi_values = np.array([0, np.pi/2, np.pi, 3*np.pi/2])
phi_dense = np.linspace(0, 2*np.pi, 500)

# z_i functions at those phi values
z1 = 2 + np.exp(1j * phi_values)
z2 = 3 / (2 - np.exp(1j * phi_values))
sigma = -0.5
omega = 1
z3 = np.exp((sigma + 1j * omega) * phi_values)

# velocities
v1 = 1j * np.exp(1j * phi_values)
v2 = 3 * 1j * np.exp(1j * phi_values) / (2 - np.exp(1j * phi_values))**2
v3 = (sigma + 1j * omega) * np.exp((sigma + 1j * omega) * phi_values)

# dense curves for plotting
z1_dense = 2 + np.exp(1j*phi_dense)
z2_dense = 3 / (2 - np.exp(1j*phi_dense))
z3_dense = np.exp((sigma + 1j*omega)*phi_dense)

# function to plot
def plot_phasors(z_dense, z, v, color, title):
    plt.figure(figsize=(6,6))
    plt.plot(z_dense.real, z_dense.imag, color=color, label=title)
    for i in range(len(z)):
        plt.arrow(z[i].real, z[i].imag, v[i].real, v[i].imag,
                  head_width=0.1, head_length=0.15, fc=color, ec=color, length_includes_head=True)
        plt.text(z[i].real + 0.1, z[i].imag + 0.1, f"v[{i}]", color=color)
    plt.title(title)
    plt.xlabel("Re(z)")
    plt.ylabel("Im(z)")
    plt.axhline(0, color='black', linewidth=0.8)
    plt.axvline(0, color='black', linewidth=0.8)
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.legend()
    plt.show()

# plot z1
plot_phasors(z1_dense, z1, v1, "red", "z()")
```

\Im(v_2) &= \frac{3 ((2 - \cos\phi)^2 - (\sin\phi)^2)}{(5 - 4\cos\phi)^2}
\\

|v_2| &= \sqrt{\Re(v_2)^2 + \Im(v_2)^2} = \boxed{\frac{3}{5 - 4 \cos\phi}}\\
\end{align}

At the specified points:

3. $z_3(\phi) = e^{(\sigma + i \omega) \phi}, \ \sigma = -0.5, \ \omega = 1$

\begin{align}
v_3(\phi) &= \frac{dz_3}{d\phi} = (\sigma + i \omega) e^{(\sigma + i \omega) \phi} \\

\end{align}

At the specified points:

\end{align*}

**Using your result from Question 5B, find the intensity $I$ incident on the mirror.**

---
Intensity "represents the average power per unit area transferred by... the light, and can be found simply by taking the average value of the Poynting vector" (from lecture module 1).
$$
I = \langle \mathbf{S} \rangle = \frac{1}{T} \int_0^T \mathbf{S}(t) \, dt,
$$

where $T = 2\pi/\omega$ is the wave period.  

For the spherical-wave Poynting vector from Question 5B:
$$
\mathbf{S}(r,\theta,t) = 
\frac{A^2 \sin\theta}{\mu_0 r} 
\left( \cos u - \frac{1}{k r} \sin u \right)
\Bigg[
\left( k \cos u + \frac{1}{k r^2} \cos u + \frac{1}{r} \sin u \right) \hat{\mathbf{r}}- 
\frac{2 \cos\theta}{\omega r} \left( \sin u + \frac{1}{k r} \cos u \right) \hat{\boldsymbol{\theta}}
\Bigg],
$$

with $u = kr - \omega t$.

---
Trig identities:
$$
\cos^2 u = \frac{1 + \cos 2u}{2}, \quad \sin^2 u = \frac{1 - \cos 2u}{2}
$$

Integrated over one period:
$$
\langle \cos^2 u \rangle = \frac{1}{T} \int_0^T \frac{1 + \cos 2u}{2} \, dt = \frac{1}{2}, \quad
\langle \sin^2 u \rangle = \frac{1}{2}
$$

Cross terms average to zero:
$$
\langle \sin u \cos u \rangle = \frac{1}{2T} \int_0^T \sin 2u \, dt = 0
$$

---
Integrate, expand, and keep nonzero terms:
$$
\langle S_r \rangle 
= \frac{1}{\mu_0} \langle E_\phi B_\theta \rangle
= \frac{1}{\mu_0} E_\phi^2 \frac{1}{c} \langle \cos^2 u \rangle
= \frac{E_\phi^2}{2 \mu_0 c}.
$$

Substitute $E_\phi = A \frac{\sin\theta}{r}$:
$$
\langle S_r \rangle = \frac{A^2 \sin^2\theta}{2 \mu_0 c \, r^2}.
$$

---
All terms in $S_\theta$ involve $\sin u \cos u$ or are smaller by $1/(kr)$, so
$$
\langle S_\theta \rangle \sim 0
$$

---
Intensity:
$$
\boxed{
I(r(\theta),\theta) = \langle \mathbf{S} \rangle \approx \langle S_r \rangle = \frac{A^2 \sin^2\theta}{2 \mu_0 c \, r^2}.
}
$$

### **Question 5D:**

* **Find the total power $P$ incident on the mirror.**  
* **Compare to the total power emitted by the spherical wave.**

---
From 5C:
$$
I(r(\theta),\theta) = \frac{A^2 \sin^2 \theta}{2 \mu_0 c \, r(\theta)^2}.
$$

---
For a little piece of mirror:
$$
dP = I \, dA
$$

Spherical coordinates:
$$
dA_\perp = r^2 \, d\Omega, \quad d\Omega = \sin\theta \, d\theta \, d\phi
$$

Then:
$$
dP = I(r,\theta) \, r^2 \, d\Omega
= \frac{A^2 \sin^2 \theta}{2 \mu_0 c \, r^2} \, r^2 \sin\theta \, d\theta \, d\phi
= \frac{A^2}{2 \mu_0 c} \, \sin^3 \theta \, d\theta \, d\phi.
$$

---
Let $R$ be the mirror radius located at distance $r_0$ from the source:
* $\theta_\text{max} \approx \arctan(R / r_0)$  

Total power incident:
$$
\begin{aligned}
P_\text{mirror} &= \int_0^{2\pi} d\phi \int_0^{\theta_\text{max}} \frac{A^2}{2 \mu_0 c} \sin^3 \theta \, d\theta \\
&= \frac{A^2}{2 \mu_0 c} (2\pi) \int_0^{\theta_\text{max}} \sin^3 \theta \, d\theta \\
&= \frac{\pi A^2}{\mu_0 c} \int_0^{\theta_\text{max}} \sin^3 \theta \, d\theta
\end{aligned}
$$

---
Use the identity $\sin^3 \theta = \sin\theta - \sin\theta \cos^2\theta$:
$$
\int_0^{\theta_\text{max}} \sin^3 \theta \, d\theta
= \frac{2}{3} - \cos\theta_\text{max} + \frac{\cos^3\theta_\text{max}}{3}
$$

Total power incident:
$$
\boxed{
P_\text{mirror} = \frac{\pi A^2}{\mu_0 c} \left( \frac{2}{3} - \cos\theta_\text{max} + \frac{\cos^3\theta_\text{max}}{3} \right)
}
$$

---
For a full spherical wave, integrate over the entire solid angle $\theta \in [0,\pi]$:

$$
P_\text{total} = \int_0^{2\pi} d\phi \int_0^\pi I(r,\theta) r^2 \sin\theta \, d\theta
= \frac{A^2}{2 \mu_0 c} (2\pi) \int_0^\pi \sin^3\theta \, d\theta
= \frac{\pi A^2}{\mu_0 c} \int_0^\pi \sin^3\theta \, d\theta
$$

$$
\int_0^\pi \sin^3 \theta \, d\theta = \frac{4}{3}
$$

$$
\Rightarrow P_\text{total} = \frac{4\pi A^2}{3 \mu_0 c}
$$

---
**Comparison**  
Fraction of total power intercepted by the mirror:
$$
\frac{P_\text{mirror}}{P_\text{total}} = \frac{3}{4} \left( \frac{2}{3} - \cos\theta_\text{max} + \frac{\cos^3\theta_\text{max}}{3} \right)
$$


---
**Small-angle approximation for the mirror ($R \ll r_0$)**  
If the mirror is small or you're perhaps far away, you might be able to get away with the small angle:
$$
\theta_\text{max} \approx \frac{R}{r_0}.
$$

Then, using the same formula:
$$
P_\text{mirror} \approx \frac{\pi A^2}{\mu_0 c} \int_0^{\theta_\text{max}} \sin^3 \theta \, d\theta
$$

For $\theta_\text{max} \ll 1$, expand $\sin\theta \approx \theta$:
$$
\sin^3 \theta \approx \theta^3
$$

Then the integral becomes:
$$
\int_0^{\theta_\text{max}} \sin^3 \theta \, d\theta \approx \int_0^{\theta_\text{max}} \theta^3 \, d\theta = \frac{\theta_\text{max}^4}{4}
$$

Substitute $\theta_\text{max} \approx R / r_0$:
$$
\int_0^{\theta_\text{max}} \sin^3 \theta \, d\theta \approx \frac{1}{4} \left( \frac{R}{r_0} \right)^4
$$

Hence the **total power on the mirror**:
$$
\boxed{
P_\text{mirror} \approx \frac{\pi A^2}{4 \mu_0 c} \left( \frac{R}{r_0} \right)^4
}
$$

---
**Comparison to total spherical-wave power**:

$$
P_\text{total} = \frac{4 \pi A^2}{3 \mu_0 c}
$$

Fraction intercepted:
$$
\frac{P_\text{mirror}}{P_\text{total}} \approx \frac{3}{16} \left( \frac{R}{r_0} \right)^4 \ll 1
$$


<div style="background-color: #331c2e; padding: 3px; border-radius: 5px;">

### **Question 5E:**

</div>

* **Calculate the radiation pressure $p_\mathrm{rad}$ incident on the mirror.** 
* **Find the radiation pressure force $F_\mathrm{rad}$.**  
    * *Assume the mirror is a perfect reflector.*  

* **If the mirror has a mass $m$, what is its acceleration?**

---
For a perfectly reflecting surface, the radiation pressure is related to the intensity $I$ by:
$$
p_\mathrm{rad} = \frac{2 I}{c}
$$

where the factor of 2 accounts for reflection.

---
Use the total power incident on the mirror $P_\mathrm{mirror}$ from 5D and the mirror area $A_\mathrm{mirror}$:

$$
p_\mathrm{rad}(r,\theta) 
    = \frac{2 I(r,\theta)}{c} 
    = \frac{2}{c} \cdot \frac{A^2 \sin^2 \theta}{2 \mu_0 c \, r^2} 
    = \boxed{\frac{A^2 \sin^2 \theta}{\mu_0 c^2 \, r^2}}
$$

The total force on the mirror is:
$$
F_\mathrm{rad} 
    = p_\mathrm{rad} \, A_\mathrm{mirror} 
    = \frac{2 P_\mathrm{mirror}}{c} 
    \approx \frac{2}{c} \cdot \frac{\pi A^2}{4 \mu_0 c} \left( \frac{R}{r_0} \right)^4
    = \boxed{\frac{\pi A^2}{2 \mu_0 c^2} \left( \frac{R}{r_0} \right)^4}
$$

where $P_\mathrm{mirror} = I_\mathrm{avg} \, A_\mathrm{mirror}$.

---
If the mirror has mass $m$, its acceleration is:
$$
\boxed{
a =\frac{F_\mathrm{rad}}{m} \approx \frac{\pi A^2}{2 \mu_0 c^2 m} \left( \frac{R}{r_0} \right)^4
}
$$
