### End Mirror Modulation
Now, suppose we apply some common modulation to both end mirrors:
$$\phi_c(t) = \phi_x(t) = \phi_y(t) = \Gamma \cos(\omega t)$$
Write what the new arm reflected transfer functions are for:
$$\dfrac{E_\mathrm{x}}{E_\mathrm{in}}, \dfrac{E_\mathrm{y}}{E_\mathrm{in}}$$
*Hint: Remember that you should end up with three distinct fields at $\omega_0$ and $\omega_0 \pm \omega$* 

---

When you apply a common phase modulation to both arms, the round-trip phase becomes

$$
2\phi_{x,y}(t) = 2\phi_{x,y}^{(0)} + 2\Gamma \cos(\omega t),
$$

where $\phi_{x,y}^{(0)} = \omega_0 L_{x,y}$.


Using
$$
e^{i 2\Gamma \cos(\omega t)} \approx1 + i 2\Gamma \cos(\omega t) 
\quad (\Gamma \ll 1),
$$

and

$$
\cos(\omega t)=\frac{1}{2}\left(e^{i\omega t} + e^{-i\omega t}\right),
$$

we obtain

$$
e^{i 2\Gamma \cos(\omega t)}\approx1+i\Gamma\left(e^{i\omega t} + e^{-i\omega t}\right).
$$

This produces three frequency components:
- carrier at $\omega_0$,
- upper sideband at $\omega_0 + \omega$,
- lower sideband at $\omega_0 - \omega$.


**Reflected Arm Transfer Functions:**  
* X-arm:

$$
\boxed{
\frac{E_x}{E_\mathrm{in}}=
t_\mathrm{bs} r_x e^{2i\phi_x^{(0)}}\left[1+i\Gamma e^{i\omega t}+i\Gamma e^{-i\omega t}\right]
}
$$

Y-arm
$$
\boxed{
\frac{E_y}{E_\mathrm{in}}=
r_\mathrm{bs} r_y e^{2i\phi_y^{(0)}}\left[1+i\Gamma e^{i\omega t}+\Gamma e^{-i\omega t}\right]
}
$$


Each reflected arm field now contains:

- A carrier at $\omega_0$,
- An upper sideband at $\omega_0 + \omega$,
- A lower sideband at $\omega_0 - \omega$,

with sideband amplitude proportional to $\Gamma$.

### Propagate to the antisymmetric port
Propagate your new transfer functions to the antisymmetric port, 
and compute a new $\dfrac{E_\mathrm{as}}{E_\mathrm{in}}$.

Be careful here.  
Consider carefully what $k = \omega/c$ is for each of your electric fields as they propagate back to the beamsplitter, accruing some phase $\phi = -k L$.\
Should each field accrue the same phase?

---
The sidebands do **not** accumulate the same phase as the carrier when propagating a distance $L$:
$$
k_0 = \frac{\omega_0}{c},
\qquad
k_\pm = \frac{\omega_0 \pm \omega}{c}.
$$

Define static round-trip phase:
$$
\Phi_{x,y}(\Omega) = 2 \frac{\Omega L_{x,y}}{c}.
$$

Then the X-arm field becomes:
$$
\frac{E_x}{E_\mathrm{in}}=
t_\mathrm{bs} r_x\left[e^{i\Phi_x(\omega_0)}+i\Gamma e^{i\Phi_x(\omega_0+\omega)}+i\Gamma e^{i\Phi_x(\omega_0-\omega)}\right].
$$

Y-arm:
$$
\frac{E_y}{E_\mathrm{in}}=
r_\mathrm{bs} r_y\left[e^{i\Phi_y(\omega_0)}+i\Gamma e^{i\Phi_y(\omega_0+\omega)}+i\Gamma e^{i\Phi_y(\omega_0-\omega)}\right].
$$


Recombining at the beamsplitter:

$$
E_\mathrm{as}=
t_\mathrm{bs} E_x-r_\mathrm{bs} E_y.
$$

Therefore,
$$
\frac{E_\mathrm{as}}{E_\mathrm{in}}=
t_\mathrm{bs}^2 r_x\left[e^{i\Phi_x(\omega_0)}+i\Gamma e^{i\Phi_x(\omega_0+\omega)}+i\Gamma e^{i\Phi_x(\omega_0-\omega)}\right]-r_\mathrm{bs}^2 r_y\left[e^{i\Phi_y(\omega_0)}
+i\Gamma e^{i\Phi_y(\omega_0+\omega)}
+i\Gamma e^{i\Phi_y(\omega_0-\omega)}
\right].
$$

Each frequency component accumulates a phase
$$
\phi = - k L = -\frac{\Omega L}{c},
$$

with its **own** $\Omega$.

So:

- The carrier interferes according to $e^{2i\omega_0 L/c}$.
- The upper sideband interferes according to $e^{2i(\omega_0+\omega)L/c}$.
- The lower sideband interferes according to $e^{2i(\omega_0-\omega)L/c}$.

Because $L_x \neq L_y$, the sidebands experience a **frequency-dependent differential phase**, producing the full Michelson frequency response.

The fields therefore do **not** accrue the same phase.
