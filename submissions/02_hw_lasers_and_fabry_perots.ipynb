## **Geometric Series Fabry-Perot**

### **Part A:** 
Rederive the Fabry-Perot intracavity electric field $E_\mathrm{cav}$ using the fact that the infinite geometric series
$$
\sum_{n=0}^\infty x^n = 1 + x + x^2 + \cdots x^n + \cdots = \dfrac{1}{1 - x} \qquad \text{iff } |x| < 1
$$

---

Let:

- $E_0$ = incident electric field  
- $t_1$ = amplitude transmission coefficient of mirror 1  
- $r_1, r_2$ = amplitude reflection coefficients  
- $L$ = cavity length  
- $k = \dfrac{2\pi}{\lambda}$  

The phase accumulated in one round trip is:
$$
\phi = 2kL
$$

---
**First Transmitted Field Inside the Cavity:**

The first transmitted field is:
$$
E_1 = t_1 E_0
$$

After one round trip inside the cavity, the field picks up:

- Reflection factor: $r_1 r_2$
- Phase factor: $e^{i\phi}$

So each successive circulating field gains a factor:
$$
r_1 r_2 e^{i\phi}
$$

---

**Infinite Sum of All Circulating Fields:**

The total intracavity field is the sum of:
$$
E_{\text{cav}} = t_1 E_0 \left[1 + (r_1 r_2 e^{i\phi})+ (r_1 r_2 e^{i\phi})^2+ \cdots\right]
$$

This is a geometric series with:
$$
x = r_1 r_2 e^{i\phi}
$$

Given that:
$$
|r_1 r_2| < 1
$$

we can use:
$$
\sum_{n=0}^{\infty} x^n = \frac{1}{1-x}
$$

---

**Final Expression for $\mathbf{E_{\textbf{cav}}}$:**

$$
E_{\text{cav}} = t_1 E_0 \frac{1}{1 - r_1 r_2 e^{i\phi}}
$$

So $E_{\text{cav}}$ is:
$$
E_{\text{cav}} = t_1 E_0 \frac{1}{1 - r_1 r_2 e^{i\phi}}
  = \boxed{\frac{t_1 E_0}{1 - r_1 r_2 e^{i 2kL}}}
$$

### **Part B:**
Draw a plot of the first couple of electric fields $E_n$,
as well as the total phasor $E_\mathrm{cav}$,
1. while on resonance,
2. while just off resonance, $\phi_\mathrm{rt} \neq 0$, but $\phi_\mathrm{rt} \ll 1$.

```python
plt.close('all')
# parameters (I just kind of chose these)
r1 = 0.9
r2 = 0.9
E0 = 1.0
t1 = np.sqrt(1 - r1**2)
N_terms = 10

def plot_phasors(phi_rt, title):
    x = r1 * r2 * np.exp(1j * phi_rt)
    En = np.array([t1 * E0 * x**n for n in range(N_terms)])
    Ecav = np.sum(En)
    
    plt.figure()
    cumulative = 0 + 0j

    # colormap for individual E_n
    colors = plt.cm.viridis(np.linspace(0, 1, N_terms))

    # keep track of cumulative endpoints for axis scaling
    cumulative_points = []
    
    # plot each E_n head-to-tail
    for n, (e, c) in enumerate(zip(En, colors), start=1):
        plt.quiver(cumulative.real,
                   cumulative.imag,
                   e.real,
                   e.imag,
                   angles='xy',
                   scale_units='xy',
                   scale=1,
                   color=c,
                   label=f"$E_{n}$")
        cumulative += e
        cumulative_points.append(cumulative)

    cumulative_points = np.array(cumulative_points)
    
    # plot total E_cav
    plt.quiver(0, 0,
               Ecav.real,
               Ecav.imag,
               angles='xy',
               scale_units='xy',
               scale=1,
               color='red',
               width=0.005,
               label="$E_{\\rm cav}$")

    # set axis limits
    all_points_real = np.concatenate(([0], cumulative_points.real, [Ecav.real]))
    all_points_imag = np.concatenate(([0], cumulative_points.imag, [Ecav.imag]))

    x_min, x_max = np.min(all_points_real), np.max(all_points_real)
    y_min, y_max = np.min(all_points_imag), np.max(all_points_imag)

    x_margin = 0.2 * (x_max - x_min) if (x_max - x_min) != 0 else 0.1
    y_margin = 0.2 * (y_max - y_min) if (y_max - y_min) != 0 else 0.1

    plt.xlim(x_min - x_margin, x_max + x_margin)
    plt.ylim(y_min - y_margin, y_max + y_margin)
    plt.axhline(0, color='white', linewidth=0.5)
    plt.axvline(0, color='white', linewidth=0.5)
    plt.gca().set_aspect('equal', 'box')
    plt.xlabel("Re(E)")
    plt.ylabel("Im(E)")
    plt.title(title)
    plt.legend(loc='upper left', fontsize=8)
    plt.show()


# On resonance
plot_phasors(phi_rt=0.0, title="Phasor Buildup (On Resonance)")

# Slightly off resonance
plot_phasors(phi_rt=0.2, title="Phasor Buildup (Slightly Off Resonance)")
```

1. **Round-trip time delay:**  
The round-trip time is the time for light to travel forward and back between the mirrors:
$$

$$


---
2. **Time delay for the $n$th term:**  

Each following circulating field $E_n$ contributes after one more round-trip:
$$
t_n = n \, \tau_{\rm rt}
$$

Or, the number of terms $n$ contributing at time $t$ is:

$$
n = \left\lfloor \frac{t}{\tau_{\rm rt}} \right\rfloor
$$

---
3. **Partial sum for $E_{\rm cav}(n)$**  

The intracavity field after $n$ terms (geometric series) is:
$$
E_{\rm cav}(n) 
    = t_1 E_{\rm in} \sum_{m=0}^{n} (r_1 r_2)^m
    = t_1 E_{\rm in} \frac{1 - (r_1 r_2)^{n+1}}{1 - r_1 r_2}
$$

---

4. **Exponential Buildup:** 

Change the form of $(r_1 r_2)^n$:  
$$
(r_1 r_2)^n \approx \exp(- n \, [ -\ln(r_1 r_2) ]) = \exp(- t / \tau_{\rm rt} -\ln(r_1 r_2))
$$

If $\tau_{\rm storage}=\tau_{\rm rt}/-\ln(r_1 r_2)$, then:

$$
E_{\rm cav}(t) \approx E_{\rm steady} \left(1 - e^{-t / \tau_{\rm storage}}\right)
$$

where $E_{\rm steady} = t_1 E_{\rm in} / (1 - r_1 r_2)$ is the steady-state intracavity field.


Then, storage time is related to the round-trip time and mirror reflectivities:
$$
\tau_{\rm storage} = \frac{\tau_{\rm rt}}{-\ln(r_1 r_2)} = \frac{2 L}{c}\frac{1}{-\ln(r_1 r_2)}
$$


---

5. **Comparison to cavity pole:**  
 
The **cavity pole** frequency $\nu_p$ is the frequency at which the intracavity field amplitude drops to $1/\sqrt{2}$ for a sinusoidal modulation of the input.  
From chapter 3, we can use the transfer function to find the pole:

$$
1 &- r_1 r_2 e^{i 2 \omega_p L / c} = 0\\
\rightarrow i \omega_p &=  \dfrac{c}{2 L} -\ln\left(r_1 r_2\right)\\
i \nu_p &= \dfrac{1}{2 \pi} \dfrac{c}{2 L} -\ln\left(r_1 r_2\right)
$$

This is just the inverse of the storage time... almost:
$$
\nu_p = \frac{1}{2 \pi \tau_{\rm storage}} = \frac{-\ln(r_1 r_2)}{2 \pi \, \tau_{\rm rt}}
$$

## **Finesse and Loss in a Fabry-Perot**
A very convenient relationship between total loss in a cavity $\mathcal{L}_\mathrm{total}$ 
and cavity finesse $\mathcal{F}$ can be calculated to be (in the high-finesse limit $\mathcal{F} \gg 1$):
$$
\mathcal{F} = \dfrac{2 \pi}{\mathcal{L}_\mathrm{total}}
$$

Derive this result starting with $\mathcal{F} = \dfrac{\mathrm{FSR}}{\nu_\mathrm{FWHM}}$.

*Hint 1:* Total loss includes transmission losses for both mirrors: $\mathcal{L}_\mathrm{total} = T_1 + T_2 + \mathcal{L}_1 + \mathcal{L}_2$.

*Hint 2:* Write $r = \sqrt{1 - T - \mathcal{L}}$.

*Hint 3:* Use the binomial approximation.

*Hint 4:* This paper from MIT may be helpful: [*Loss in long-storage-time optical cavities*](https://arxiv.org/abs/1310.1820)

---


Goal: Derive the relationship between the finesse $\mathcal{F}$ and the total round-trip loss $\mathcal{L}_{\rm total}$ in the **high-finesse limit** ($\mathcal{F} \gg 1$).

---
1. **Finesse definition**

$$
\mathcal{F} = \frac{\mathrm{FSR}}{\nu_{\rm FWHM}}
$$

where:

- $\mathrm{FSR} = c / (2L)$ is the **free spectral range**,  
- $\nu_{\rm FWHM}$ is the **full width at half maximum** of a cavity resonance.

---

2. **Transmission coefficient and total loss**

For a cavity with mirror transmissions $T_1, T_2$ and additional losses $\mathcal{L}_1, \mathcal{L}_2$, the **total round-trip loss** is:

$$
\mathcal{L}_{\rm total} = T_1 + T_2 + \mathcal{L}_1 + \mathcal{L}_2
$$

Field reflection coefficient:
$$
r_i = \sqrt{1 - T_i - \mathcal{L}_i}
$$

Total round-trip field reflection (ignoring phase):
$$
r_1 r_2 \approx \sqrt{(1 - T_1 - \mathcal{L}_1)(1 - T_2 - \mathcal{L}_2)}
$$

---

3. **High-finesse approximation**

In high-finesse cavities, losses are small:
$$
T_i, \mathcal{L}_i \ll 1
$$

We can use the binomial approximation:
$$
\sqrt{1 - \epsilon} \approx 1 - \frac{\epsilon}{2}, \quad \epsilon \ll 1
$$

Then:
$$
r_1 r_2 \approx 1 - \frac{T_1 + T_2 + \mathcal{L}_1 + \mathcal{L}_2}{2} = 1 - \frac{\mathcal{L}_{\rm total}}{2}
$$

---
4. **Relate finesse to mirror reflection**

The FWHM of a resonant peak is related to the field reflection by:
$$
\nu_{\rm FWHM} = \frac{\mathrm{FSR}}{\pi} \arccos(r_1 r_2)
$$

* For $r_1 r_2 \approx 1$, we can approximate:
$$
\arccos(r_1 r_2) \approx \sqrt{2 (1 - r_1 r_2)}
$$

* Using $1 - r_1 r_2 \approx \mathcal{L}_{\rm total}/2$:
$$
\nu_{\rm FWHM} \approx \frac{\mathrm{FSR}}{\pi} \sqrt{ \mathcal{L}_{\rm total} }
$$

* In the small-loss limit, I think you can linearize:
$$
\nu_{\rm FWHM} \approx \frac{\mathrm{FSR}}{2 \pi} \, \mathcal{L}_{\rm total}
$$

---

5. **Finesse in terms of total loss**

Using the definition:
$$
\mathcal{F} = \frac{\mathrm{FSR}}{\nu_{\rm FWHM}} \approx \frac{\mathrm{FSR}}{\mathrm{FSR} \, \mathcal{L}_{\rm total} / (2\pi)} 
    = \boxed{\frac{2 \pi}{\mathcal{L}_{\rm total}}}
$$

---

## **Finesse and Gain in a Fabry-Perot**


### **Part A:**
Assuming that we have a *critically-coupled* Fabry-Perot cavity,\
what is the comparison between the finesse $\mathcal{F}$ and power gain $G_\mathrm{cav}$?

---
1. **Cavity power gain**

The steady-state intracavity power relative to the input power is:

$$
G_{\rm cav} = \frac{P_{\rm cav}}{P_{\rm in}} = \frac{T_1}{(1 - \sqrt{1 - \mathcal{L}_{\rm total}})^2}
$$

where:
- $T_1$ is the transmission of the input mirror,  
- $\mathcal{L}_{\rm total}$ is the total round-trip loss (including the second mirror and any internal losses).

For high-finesse cavities ($\mathcal{L}_{\rm total} \ll 1$), we can use the approximation:
$$
\sqrt{1 - \mathcal{L}_{\rm total}} \approx 1 - \frac{\mathcal{L}_{\rm total}}{2}
$$

Then the power gain simplifies to:
$$
G_{\rm cav} \approx \frac{T_1}{(\mathcal{L}_{\rm total}/2)^2} = \frac{4 T_1}{\mathcal{L}_{\rm total}^2}
$$

---
2. **Cavity finesse**

High-finesse result from problem #4:
$$
\mathcal{F} = \frac{2 \pi}{\mathcal{L}_{\rm total}}
$$

---
3. **Relation between finesse and power gain**

For a critically-coupled cavity, the input mirror transmission is chosen to match the internal losses:

$$
T_1 = \mathcal{L}_{\rm total} / 2
$$

Substitute into the power gain expression:
$$
G_{\rm cav} \approx \frac{4 ( \mathcal{L}_{\rm total} / 2 )}{\mathcal{L}_{\rm total}^2} = \frac{2}{\mathcal{L}_{\rm total}} 
$$

Now, express $\mathcal{L}_{\rm total}$ in terms of finesse:
$$
\mathcal{L}_{\rm total} = \frac{2 \pi}{\mathcal{F}}
$$

Power gain in terms of finesse is now:
$$
\boxed{
G_{\rm cav} \approx \frac{2}{2 \pi / \mathcal{F}} = \frac{\mathcal{F}}{\pi}
}
$$

---

### **Part B:**
Repeat Part A above for an *over-coupled* cavity such that $r_2 \approx 1$.\
Does the relationship change?

---
New condition:
$$
r_2 \approx 1 \quad \Rightarrow \quad T_2 \approx 0
$$

Also assume internal losses are small ($\mathcal{L}_1 + \mathcal{L}_2 \ll 1$).

---

1. **Cavity power gain**

The intracavity power gain is still:
$$
G_{\rm cav} = \frac{T_1}{(1 - \sqrt{r_1 r_2})^2} \approx \frac{T_1}{(1 - \sqrt{1 - T_1 - \mathcal{L}_{\rm int}})^2}
$$

where $\mathcal{L}_{\rm int} = \mathcal{L}_1 + \mathcal{L}_2$ is the internal round-trip loss.  

- Approximating for high reflectivity:  
$$
\sqrt{1 - T_1 - \mathcal{L}_{\rm int}} \approx 1 - \frac{T_1 + \mathcal{L}_{\rm int}}{2}
$$

Then:
$$
G_{\rm cav} \approx \frac{T_1}{\left( \frac{T_1 + \mathcal{L}_{\rm int}}{2} \right)^2} = \frac{4 T_1}{(T_1 + \mathcal{L}_{\rm int})^2}
$$

---
2. **Finesse**

The finesse is still:
$$
\mathcal{F} = \frac{2 \pi}{\mathcal{L}_{\rm total}} \approx \frac{2 \pi}{T_1 + \mathcal{L}_{\rm int}}
$$

since $T_2 \approx 0$.

---
3. **Relation between finesse and power gain**

For over-coupled cavities, the input mirror transmission is larger than the internal losses:
$$
T_1 \gg \mathcal{L}_{\rm int}
$$

- Then the cavity gain becomes:
$$
G_{\rm cav} \approx \frac{4 T_1}{T_1^2} = \frac{4}{T_1}
$$

- Expressing $T_1$ in terms of finesse:
$$
T_1 \approx \mathcal{L}_{\rm total} \approx \frac{2 \pi}{\mathcal{F}}
$$

- So the gain becomes:
$$
\boxed{
G_{\rm cav} \approx \frac{4}{2 \pi / \mathcal{F}} = \frac{2 \, \mathcal{F}}{\pi}
}
$$

Compared to the critically-coupled case, the intracavity power gain is ~twice as large for the same finesse. The input mirror dominates the losses, and the cavity stores more power before internal losses limit the buildup.


## **Reflection Phase Angle vs. Frequency**  

***Reflection phase angle versus frequency.*** Compute and plot the phase angle versus frequency for the complex interferometer reflectivity $\tilde{E}_{refl}/\tilde{E}_{inc}$ for some typical choices of $R_1$, $R_2$, and $\delta_0$. Consider in particular the case of the *Gires-Tounois interferometer*, which has $R_1$ finite, $R_2=100\%$ and $\delta_0$, so that the magnitude of the overall reflectivity is constant and equal to 1 at all frequencies, but the phase angle of the reflection changes sharply with frequency. (Hint: Consider the phase angles versus frequncy for the numerator and denominator separately, then combine.

---

* **Complex reflectivity of a FabryPerot cavity**

The complex field reflectivity is

$$
r(\omega) = \frac{r_1 - r_2 e^{i 2 \delta(\omega)}}{1 - r_1 r_2 e^{i 2 \delta(\omega)}}
$$

where:
- $r_1, r_2$ are field reflection coefficients
- $\delta(\omega) = \frac{\omega L}{c} + \delta_0$ is the single-pass phase
- $2\delta(\omega)$ is the round-trip phase

Phase angle:
$$
\phi(\omega) = \arg \left[ r(\omega) \right].
$$

Write the reflectivity as:
$$
r(\omega) = \frac{N(\omega)}{D(\omega)}
$$

* where $N(\omega) = r_1 - r_2 e^{i 2\delta}$ and $D(\omega) = 1 - r_1 r_2 e^{i 2\delta}$

Then the phase is
$$
\phi(\omega) = \arg[N(\omega)] - \arg[D(\omega)].
$$


---
For a **GiresTournois interferometer**:

$$
R_2 = 1 \quad \Rightarrow \quad r_2 = 1
$$

and $R_1$ is finite.

The reflectivity becomes
$$
r(\omega) = \frac{r_1 - e^{i 2\delta}}{1 - r_1 e^{i 2\delta}}.
$$

Compute the magnitude:

$$
|r(\omega)|^2 = \frac{|r_1 - e^{i 2\delta}|^2}{|1 - r_1 e^{i 2\delta}|^2}.
$$

Expanding the numerator and denominator gives the same expression:
$$
|r(\omega)|^2 = 1.
$$

Thus, at all frequencies:
$$
\boxed{
|r(\omega)| = 1 \quad \text{for all frequencies}
}
$$

The interferometer reflects all power, but with a **frequency-dependent phase shift**.

---
* **Phase Behavior:**

Since the magnitude is constant, all the interesting behavior is in

$$
\phi(\omega) = \arg(r_1 - e^{i 2\delta})-\arg(1 - r_1 e^{i 2\delta}).
$$

Near resonance:
$$
2\delta(\omega) \approx 2\pi m
$$

the denominator becomes small in phase slope, which makes the phase change quickly.

### **Plots**  
Make Bode plots of the reflection phase vs round-trip phase where $\phi$ serves as the x-axis.\
*Hint:* This is similar to our expressions $\theta(\phi)$, where $\theta$ is the reflection phase, and $\phi$ is the round-trip phase.

```python
plt.close('all')
from matplotlib.widgets import Slider

# phase ranges
phi_fp  = np.linspace(-np.pi, np.pi, 4000)
phi_gti = np.linspace(-3*np.pi, 3*np.pi, 8000)  # wider range to see more of what it does

def reflection_coefficient(r1, r2, phi):
    return (r1 - r2*np.exp(1j*phi)) / (1 - r1*r2*np.exp(1j*phi))

def compute_phase(R, r2, phi):
    r1 = np.sqrt(R)
    r = reflection_coefficient(r1, r2, phi)
    return np.unwrap(np.angle(r))

R_init = 0.9

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))
plt.subplots_adjust(bottom=0.25)

# initial curves
line_fp,  = ax1.plot(phi_fp,  compute_phase(R_init, np.sqrt(R_init), phi_fp))
line_gti, = ax2.plot(phi_gti, compute_phase(R_init, 1.0, phi_gti))

# formatting
ax1.set_title("Symmetric FabryPerot")
ax1.set_xlabel("Round-trip phase  (rad)")
ax1.set_ylabel("Reflection phase () (rad)")
ax1.set_xlim(-np.pi, np.pi)
ax1.grid(True)

ax2.set_title("GiresTournois (R = 1)")
ax2.set_xlabel("Round-trip phase  (rad)")
ax2.set_xlim(-3*np.pi, 3*np.pi)
ax2.grid(True)

# slider
ax_R = plt.axes([0.25, 0.1, 0.5, 0.03])
slider_R = Slider(ax_R, "Reflectivity R", 0.5, 0.999, valinit=R_init)

def update(val):
    R = slider_R.val
    
    line_fp.set_ydata(compute_phase(R, np.sqrt(R), phi_fp))
    line_gti.set_ydata(compute_phase(R, 1.0, phi_gti))
    
    fig.canvas.draw_idle()

slider_R.on_changed(update)

plt.show()
```
